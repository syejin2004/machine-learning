```python
import pandas as pd
import pickle
import seaborn as sns

with open('fifa_test.pkl', 'rb') as train_data:
  train = pickle.load(train_data)

with open('fifa_train.pkl', 'rb') as test_data:
  test = pickle.load(test_data)

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

import numpy as np
```

```python
def svc_param_selection(X, y, nfolds):
  svm_parameters = [
      {'kernel': ['rbf'],
       'gamma': [0.1, 1],
       'C': [1, 10]
       }
  ]

  #clf = GridSearchCV(SVC(), svm_parameters, cv=nfolds)
  clf = SVC(kernel="rbf", gamma=0.0001, C=10)  #최적의 파라미터
  clf.fit(X, y)
  #print(clf.best_params_)

  return clf
```

```python
X_train = train.iloc[:, 3:24]
y_train = train[['Position']]
clf = svc_param_selection(X_train, y_train.values.ravel(), 10)
```

```python
#@title 기본 제목 테스트
#테스트에 사용될 특징을 지정
X_test = test.iloc[:, 3:24]

#특징으로 예측할 값(포지션)을 지정
y_test = test[['Position']]

#최적의 파라미터로 완성된 SVM에 테스트 데이터를 주입하여, 실제값과 예측값을 얻는다
y_true, y_pred = y_test, clf.predict(X_test)

print("accuracy" + str(accuracy_score(y_true, y_pred)))
```

```python
#실제값(ground truth)과 예측값(prediction)이 어느 정도 일치하는지 눈으로 직접 비교
comparison = pd.DataFrame({'prediction':y_pred, 'ground_truth':y_true.values.ravel()})
comparison
```

```python
comparison.to_csv('result.csv')
```
