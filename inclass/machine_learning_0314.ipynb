### 산점도 시각화 및 사이킷런을 이용한 선형 모델 훈련과 예측

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat["Life satisfaction"].values

lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()

model = LinearRegression()

model.fit(X, y)

X_new = [[37_655.2]]
print(model.predict(X_new))
```

### 택시요금 예측과 유사한 선형 데이터를 통한 머신러닝 문제

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv("data.csv")
X = lifesat[["dist"]].values
y = lifesat[["won"]].values

lifesat.plot(kind="scatter", grid=True, x="dist", y="won")
plt.axis([0, 12, 0, 35000])
plt.show()

model = LinearRegression()

model.fit(X, y)

X_new = [[8]]
print(model.predict(X_new))
```
